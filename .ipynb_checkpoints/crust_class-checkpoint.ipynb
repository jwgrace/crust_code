```python
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt
from scipy import optimize
from scipy import constants
import seaborn as sns
import time
```

```python
def pressure_to_baryon_density(electron_density, P_tot, Z, N, units='cgs'):
    '''
    Calculates the pressure of an ion lattice given a density of background 
    electrons. 
    
    Assumes the system to be charge-neutral and that the electrons form a
    uniform, degenerate gas.
    '''
    
    pi = np.pi
    
    # Set constants according to assigned unit system
    if units == 'cgs':
        m_e = 9.1093837015e-28
        c = 2.99792458e10
        m_ec2 = m_e*c**2
        h_bar = 1.054571817e-27
        h_barc = h_bar*c
        alpha = 0.0072973525693
    
    elif units == 'nuclear':
        m_ec2 = 0.5110
        h_bar = 6.582e-22
        h_barc = 197.3
        alpha = 0.0072973525693
    
    # Electron relativity parameter (unitless)
    x_r = h_barc*(3*pi**2*electron_density)**(1/3)/m_ec2

    # Electron energy density (MeV/fm^3)
    electron_energy_density = m_ec2**4/(8*pi**2*h_barc**3)*(x_r*(2*x_r**2 + 1)*np.sqrt(x_r**2 + 1) - np.log(x_r + np.sqrt(x_r**2 + 1)))

    # Electron chemical potential (MeV)
    mu_e = m_ec2*np.sqrt(1 + x_r**2)

    # Lattice energy density (MeV/fm^3)
    lattice_energy_density = -.9*(4*pi/3)**(1/3)*Z**(2/3)*alpha*h_barc*electron_density**(4/3)

    # Electron pressure (cgs: erg/cm^3; nuclear: MeV/fm^3)
    P_e = -electron_energy_density + mu_e*electron_density
    
    return P_e + 1/3*lattice_energy_density - P_tot
```

```python
class NuclearMassTable:
    '''Table of masses of nuclides'''
    
    def __init__(self, experimental_mass_data_file, theoretical_mass_data_file):
        # Data from 'mass16-reduced.txt' and 'frdm12.dat'.
        # 'mass16-reduced.txt' contains experimental data from the 2016 Atomic Mass Evaluation.
        # 'frdm12.dat' contains mass estimates from the FRDM2012 mass model.

        # Load AME data.
        neutron_number_ame, proton_number_ame, atomic_mass_ame = np.loadtxt(experimental_mass_data_file, unpack=True)
        # Convert proton and nucleon numbers to integers so they can be used as indices.
        proton_number_ame, neutron_number_ame = proton_number_ame.astype(int), neutron_number_ame.astype(int)

        # Atomic mass unit in units of MeV/c^2
        amu = 931.49410242
        # Electron mass in units of MeV/c^2
        electron_mass = 5.48579909070e-4*amu

        # Convert atomic masses from amu to MeV/c^2 and convert to nuclear masses by subtracting electron rest mass.
        nuclear_mass_ame = atomic_mass_ame*amu - proton_number_ame*electron_mass

        # Load FRDM data
        proton_number_frdm, nucleon_number_frdm, mass_excess_frdm = np.loadtxt(theoretical_mass_data_file, unpack=True)
        # Convert proton and nucleon numbers to integers so they can be used as indices.
        proton_number_frdm, nucleon_number_frdm = proton_number_frdm.astype(int), nucleon_number_frdm.astype(int)
        # Calculate neutron number.
        neutron_number_frdm = nucleon_number_frdm - proton_number_frdm

        # Nuclear masses from FRDM data.
        nuclear_mass_frdm = mass_excess_frdm + nucleon_number_frdm*amu - proton_number_frdm*electron_mass

        # Create 2D array of masses with axes of Z and N
        # We will take experimental (AME) values where possible. Otherwise will use model (FRDM) values.

        # Determine overall dimensions of array based on the largest values of Z and N in both datasets.
        self.Z_max = np.max([np.max(proton_number_ame), np.max(proton_number_frdm)])
        self.N_max = np.max([np.max(neutron_number_ame), np.max(neutron_number_frdm)])

        # We don't have data for all nuclides in the ZxN space, so we initialize values with NaN and will fill in the values for which we have data.
        self.nuclear_mass = np.full((int(np.max(self.Z_max)) + 1, int(np.max(self.N_max)) + 1), np.NaN)

        # First load model masses then experimental masses. Experimental values will overwrite model values where they exist.
        # Iterate through each nuclide, determine Z and N values for that nuclide, then store its nuclear mass in the appropriate space in the ZxN array.

        # FRDM model masses
        for i, Z in enumerate(proton_number_frdm):
            N = neutron_number_frdm[i]

            self.nuclear_mass[Z, N] = nuclear_mass_frdm[i]

        # AME experimental masses
        for i, Z in enumerate(proton_number_ame):
            N = neutron_number_ame[i]

            self.nuclear_mass[Z, N] = nuclear_mass_ame[i]
```

```python
class CrustCell:
    '''A volume element cell of a neutron star crust'''
    
    def __init__(self, total_pressure, experimental_mass_data_file, theoretical_mass_data_file):
        
        nuclear_mass_table = NuclearMassTable(experimental_mass_data_file, theoretical_mass_data_file)
        
        self.Z_max = nuclear_mass_table.Z_max
        self.N_max = nuclear_mass_table.N_max
        
        pi = np.pi
        MeV_to_erg = 1.60218e-6

        units = 'cgs'

        # Set constants according to assigned unit system
        if units == 'cgs':
            m_e = 9.1093837015e-28
            c = 2.99792458e10
            m_ec2 = m_e*c**2
            h_bar = 1.054571817e-27
            h_barc = h_bar*c
            alpha = 0.0072973525693

        elif units == 'nuclear':
            m_ec2 = 0.5110
            h_bar = 6.582e-22
            h_barc = 197.3
            alpha = 0.0072973525693

        # Create arrays of proton number Z, neutron number N, and nucleon number A.
        # Useful to have for array operations.
        neutron_number, proton_number = np.meshgrid(np.arange(self.N_max + 1), np.arange(self.Z_max + 1))
        nucleon_number = proton_number + neutron_number
        
        # Limits for the baryon number density in (1/cm^3)
        # Chosen arbitarily
        n_b_min = 1e6/1.66e-24
        n_b_max = 1e12/1.66e-24

        # Minimum and maximum bounds for electron density to be used for root finding
        n_e_min = n_b_min/np.max(nucleon_number)
        n_e_max = n_b_max

        # Array of densities for each nuclide at the given pressure
        self.baryon_density = np.full_like(nuclear_mass_table.nuclear_mass, np.NaN)

        self.electron_density = np.full_like(nuclear_mass_table.nuclear_mass, np.NaN)

        # Loop through all nuclides to solve for the baryon density at the given pressure.
        for Z in range(1, self.Z_max + 1):
            for N in range(self.N_max + 1):

                # If the value for the nuclear mass is NaN, then we have no data and so we skip this nuclide.
                # If the nuclide contains no protons, it cannot form a lattice and exist as a species in the outer crust.
                if np.isnan(nuclear_mass_table.nuclear_mass[Z, N]):
                    continue

                else:
                    # Calculate baryon density using root finding method
                    self.electron_density[Z, N] = optimize.root_scalar(pressure_to_baryon_density, args=(total_pressure, Z, N), method='brentq', bracket=[n_e_min, n_e_max]).root

        self.nucleus_density = self.electron_density/proton_number
        
        # Electron density calculated from Z and A. Determined by charge neutrality
        #electron_density = baryon_density*proton_number/nucleon_number

        self.baryon_density = nucleon_number*self.nucleus_density

        # Electron relativity parameter (unitless)
        self.relativity_parameter = h_barc*(3*pi**2*self.electron_density)**(1/3)/m_ec2

        # Electron energy density (erg/cm^3)
        self.electron_energy_density = m_ec2**4/(8*pi**2*h_barc**3)*(self.relativity_parameter*(2*self.relativity_parameter**2 + 1)*np.sqrt(self.relativity_parameter**2 + 1) \
                                    - np.log(self.relativity_parameter + np.sqrt(self.relativity_parameter**2 + 1)))

        # Electron chemical potential (erg)
        self.electron_chemical_potential = m_ec2*np.sqrt(1 + self.relativity_parameter**2)

        # Electron pressure (erg/cm^3)
        self.electron_pressure = -self.electron_energy_density + self.electron_chemical_potential*self.electron_density

        # Lattice energy density (erg/cm^3)
        self.lattice_energy_density = -.9*(4*pi/3)**(1/3)*proton_number**(2/3)*alpha*h_barc*self.electron_density**(4/3)

        # Total pressure (erg/cm^3)
        self.total_pressure = self.electron_pressure + self.lattice_energy_density/3

        # Total energy density in erg/cm^3
        self.total_energy_density = self.nucleus_density*nuclear_mass_table.nuclear_mass*MeV_to_erg + self.electron_energy_density + self.lattice_energy_density

        # Total mass density in g/cm^3
        self.mass_density = self.total_energy_density/c**2

        # Gibbs free energy per nucleon in MeV
        self.gibbs_free_energy = (self.total_energy_density + total_pressure)/self.baryon_density/MeV_to_erg
        
        # Determine absolute ground state

        # The absolute ground state is the nuclide with the overall minimum value of Gibbs free energy.
        ground_index = np.nanargmin(self.gibbs_free_energy)

        # nanargmin unwraps the 2D array, so to we need to determine which 2D indices the result corresponds to.
        self.Z_ground = ground_index//(self.N_max + 1)
        self.N_ground = ground_index%(self.N_max + 1)
        
        # Determine beta stability and stability against neutron emission of each nuclide at the pressure from the previous cell

        # Arrays for each stability condition
        # Values are nan for nuclides that are unstable and will be 1 for nuclides that are stable.
        self.electron_capture_stable = np.full_like(self.gibbs_free_energy, np.NaN)
        self.electron_emission_stable = np.full_like(self.gibbs_free_energy, np.NaN)
        self.neutron_emission_stable = np.full_like(self.gibbs_free_energy, np.NaN)
        
        # Energy density and Gibbs free energy calculated for conditions in which one neutron has been emitted from each nucleus.
        # Used to determine stability against neutron emission.
        energy_density_free_neutron = self.total_energy_density + self.nucleus_density*nuclear_mass_table.nuclear_mass[0, 1]*MeV_to_erg
        baryon_density_free_neutron = (nucleon_number + 1)*self.nucleus_density
        
        self.gibbs_free_energy_free_neutron = (energy_density_free_neutron + total_pressure)/baryon_density_free_neutron/MeV_to_erg

        for Z in range(self.Z_max + 1):
            for N in range(self.N_max + 1):

                # If the value for the atomic mass is NaN, then we have no data and so we skip this nuclide.
                if np.isnan(self.gibbs_free_energy[Z, N]):
                    continue

                # Test for stability against electron capture.
                # If the current nuclide's Gibbs free energy is less than that of the product of electron capture, then it is stable.
                # If the product of electron capture has no data, we assume it is less stable than the current nuclide.
                try:
                    if np.isnan(self.gibbs_free_energy[Z-1, N+1]) or self.gibbs_free_energy[Z, N] < self.gibbs_free_energy[Z-1, N+1]:
                        self.electron_capture_stable[Z, N] = 1
                    else:
                        self.electron_capture_stable[Z, N] = 0
                except:
                    continue

                # Test for stability against electron emission.
                # If the current nuclide's Gibbs free energy is less than that of the product of electron emission, then it is stable.
                # If the product of electron emission has no data, we assume it is less stable than the current nuclide.
                try:
                    if np.isnan(self.gibbs_free_energy[Z+1, N-1]) or self.gibbs_free_energy[Z, N] < self.gibbs_free_energy[Z+1, N-1]:
                        self.electron_emission_stable[Z, N] = 1
                    else:
                        self.electron_emission_stable[Z, N] = 0
                except:
                    continue
                
                try:
                    if np.isnan(self.gibbs_free_energy_free_neutron[Z, N-1]) or self.gibbs_free_energy[Z, N] < self.gibbs_free_energy_free_neutron[Z, N-1]:
                        self.neutron_emission_stable[Z, N] = 1
                    else:
                        self.neutron_emission_stable[Z, N] = 0
                except:
                    continue
                    
        self.beta_stable = self.electron_capture_stable*self.electron_emission_stable
```

```python
experimental_mass_data_file = 'mass16-reduced.txt'
theoretical_mass_data_file = 'frdm12.dat'

P_tot = 1e28

nuclear_mass_table = NuclearMassTable(experimental_mass_data_file, theoretical_mass_data_file)
crust_cell = CrustCell(P_tot, experimental_mass_data_file, theoretical_mass_data_file)

Z_ground = crust_cell.Z_ground
N_ground = crust_cell.N_ground

gibbs_ground = crust_cell.gibbs_free_energy[Z_ground, N_ground]
mass_density_ground = crust_cell.mass_density[Z_ground, N_ground]

print('Ground State [Z, N] = [{}, {}]'.format(Z_ground, N_ground))
print('Gibbs free energy per nucleon = {}'.format(gibbs_ground))
print('Ground State denisty = {:.03e}'.format(mass_density_ground))
```

```python
beta_stable_plot = np.where(crust_cell.beta_stable == 1, 1, np.NaN)
neutron_drip_plot = np.where(crust_cell.neutron_emission_stable == 0, 1, np.NaN)

# Plot the Gibbs free energy per nucleon of each nuclide
# Highlight nuclides that are beta stable
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

plot = plt.imshow(crust_cell.gibbs_free_energy, origin='lower')
plt.imshow(beta_stable_plot, cmap='gray', origin='lower')
#plt.imshow(crust_cell.neutron_emission_stable, cmap='gray', origin='lower')
plt.imshow(neutron_drip, cmap='gray', alpha=.5, origin='lower')
#plt.imshow(neutron_capture_stable*neutron_emission_stable, cmap='gray', origin='lower')
fig.colorbar(plot)
plt.title('Gibbs Free Energy Per Nucleon (MeV)')
plt.xlabel('N')
plt.ylabel('Z')
plt.show()
```

```python
neutron_number, proton_number = np.meshgrid(np.arange(crust_cell.N_max + 1), np.arange(crust_cell.Z_max + 1))
nucleon_number = proton_number + neutron_number

A = 56

mass_mask = (nucleon_number == A) * (~np.isnan(crust_cell.gibbs_free_energy))

g_constant_A = crust_cell.gibbs_free_energy[mass_mask]
Z_constant_A = proton_number[mass_mask]
neutron_stable = crust_cell.neutron_emission_stable[mass_mask] == 1

tick_array = np.arange(Z_constant_A.min(), Z_constant_A.max() + 2, 2)

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

#plt.plot(nuclide_Z, nuclide_gibbs)
plt.plot(Z_constant_A[neutron_stable], g_constant_A[neutron_stable], '.', label='neutron stable')
plt.plot(Z_constant_A[~neutron_stable], g_constant_A[~neutron_stable], '.', label='beyond neutron drip')
plt.xlabel('Z')
plt.ylabel('Gibbs Free Energy')
plt.title('P = {}, A = {}'.format(P_tot, A))
plt.xticks(tick_array)
plt.legend()
plt.show()
```

```python
experimental_mass_data_file = 'mass16-reduced.txt'
theoretical_mass_data_file = 'frdm12.dat'

nuclear_mass_table = NuclearMassTable(experimental_mass_data_file, theoretical_mass_data_file)

num_zones = 13
# Total pressure in erg/cm^3
total_pressure = np.logspace(25, 31, num_zones)

crust_cells = []

for P_tot in total_pressure:
    crust_cells.append(CrustCell(P_tot, experimental_mass_data_file, theoretical_mass_data_file))
```

```python
neutron_number, proton_number = np.meshgrid(np.arange(crust_cell.N_max + 1), np.arange(crust_cell.Z_max + 1))
nucleon_number = proton_number + neutron_number

A_constant = 56

mass_mask = (nucleon_number == A_constant) * (~np.isnan(nuclear_mass_table.nuclear_mass))

g_constant_A = np.zeros((num_zones, np.sum(mass_mask)))
Z_constant_A = proton_number[mass_mask]
neutron_stable = np.zeros((num_zones, np.sum(mass_mask)), dtype=bool)

for i, crust_cell in enumerate(crust_cells):
    
    g_constant_A[i] = crust_cell.gibbs_free_energy[mass_mask]
    neutron_stable[i] = crust_cell.neutron_emission_stable[mass_mask] == 1
```

```python
def plot_gibbs(i):
    
    g_min = g_constant_A[i].min()
    g_min_index = np.argmin(g_constant_A[i])
    Z_min = Z_constant_A[g_min_index]
    is_neutron_stable = neutron_stable[i, g_min_index]
    
    fig = plt.figure()
    fig.patch.set_facecolor('white')
    fig.set_dpi(200)

    #plt.plot(nuclide_Z, nuclide_gibbs)
    plt.plot(Z_constant_A[neutron_stable[i]], g_constant_A[i][neutron_stable[i]], '.', color='tab:blue', label='neutron stable')
    plt.plot(Z_constant_A[~neutron_stable[i]], g_constant_A[i][~neutron_stable[i]], '.', color='tab:orange', label='beyond neutron drip')
    if is_neutron_stable:
        plt.plot(Z_min, g_min, 'x', color='tab:blue', label='minimum')
    else:
        plt.plot(Z_min, g_min, 'x', color='tab:orange', label='minimum')
    plt.ylim(gibbs_range_min, gibbs_range_max)
    plt.xlabel('Z')
    plt.ylabel('Gibbs Free Energy')
    plt.title('P = {:.3e}, A = {}'.format(total_pressure[i], A))
    plt.xticks(tick_array)
    plt.legend(loc='upper left')
    plt.show()
```

```python
%matplotlib inline
from ipywidgets import interactive

gibbs_min = g_constant_A.min()
gibbs_max = g_constant_A.max()

gibbs_range_min = gibbs_min - .125*(gibbs_max - gibbs_min)
gibbs_range_max = gibbs_max + .125*(gibbs_max - gibbs_min)

tick_array = np.arange(Z_constant_A.min(), Z_constant_A.max() + 2, 2)

interactive_plot = interactive(plot_gibbs, i=(0, num_zones-1))
output = interactive_plot.children[-1]
#output.layout.height = '350px'
interactive_plot
```
